```python
#creating a histogram with matplotlp 
plt.hist(data[["co2"]], bins= 10 )
plt.title("Distribution of Carbon Dioxide")
plt.xticks(rotation=90)
plt.xlabel("Carbon Dioxide")
plt.ylabel("Distribution")
plt.show()
```

```python
#creating a histogram for carbon dioxide with different bin sizes 
plt.hist(data[["co2"]], bins= 30 )
plt.title("Distribution of Carbon Dioxide")
plt.xticks(rotation=90)
plt.xlabel("Carbon Dioxide")
plt.ylabel("Distribution")
plt.show()
```

```python
sns.histplot(data = data, x = "feScore", bins = 30, color = "brown")
plt.xticks(rotation=90)
plt.title("Fe Score Distribution")
plt.xlabel("Fe Score")
plt.show()
```

```python
sns.histplot(data = data, x = "ghgScore", bins = 30, color = "green")
plt.xticks(rotation=90)
plt.title("Ghg Score Distribution")
plt.xlabel("Ghg Score")
plt.show()
```

```python
sns.histplot(data = data, x = "fuelType", bins = 30, color = "green")
plt.xticks(rotation=70)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
# Select numeric columns only
numeric_data = data.select_dtypes(include=["number"])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()
pallete = sns.diverging_palette(20, 220, as_cmap=True)

# Create the heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix,  annot=False, cmap=pallete,  cbar=True)
plt.title("Heatmap of Correlation Matrix ")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


file_path = 'fuel_econ.csv'  

df = pd.read_csv(file_path)


print(df.head())


plt.figure(figsize=(8, 6))
plt.hist(df['co2'], bins=15, color='#1f6feb', edgecolor='#1f6feb')
plt.title('Distribution of Carbon Dioxide Emissions')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```
