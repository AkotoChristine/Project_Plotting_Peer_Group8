```python
#creating a histogram with matplotlp 
plt.hist(data[["co2"]], bins= 10 )
plt.title("Distribution of Carbon Dioxide")
plt.xticks(rotation=90)
plt.xlabel("Carbon Dioxide")
plt.ylabel("Distribution")
plt.show()
```

```python
#creating a histogram for carbon dioxide with different bin sizes 
plt.hist(data[["co2"]], bins= 30 )
plt.title("Distribution of Carbon Dioxide")
plt.xticks(rotation=90)
plt.xlabel("Carbon Dioxide")
plt.ylabel("Distribution")
plt.show()
```

```python
sns.histplot(data = data, x = "feScore", bins = 30, color = "brown")
plt.xticks(rotation=90)
plt.title("Fe Score Distribution")
plt.xlabel("Fe Score")
plt.show()
```

```python
sns.histplot(data = data, x = "ghgScore", bins = 30, color = "green")
plt.xticks(rotation=90)
plt.title("Ghg Score Distribution")
plt.xlabel("Ghg Score")
plt.show()
```

```python
sns.histplot(data = data, x = "fuelType", bins = 30, color = "green")
plt.xticks(rotation=70)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
# Select numeric columns only
numeric_data = data.select_dtypes(include=["number"])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()
pallete = sns.diverging_palette(20, 220, as_cmap=True)

# Create the heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix,  annot=False, cmap=pallete,  cbar=True)
plt.title("Heatmap of Correlation Matrix ")
plt.show()

```

### Intepretation of CO2 Diagram

#### Frequency Analysis:

#### Bin Size Impact:
- With **fewer bins (10)**, the distribution appears smoother and highlights the general bell-shaped curve.

#### Observations:

- Extreme outliers with unusually high or low emissions are minimal, making the distribution largely unimodal and centered.

# Intepretation of Heatmap Diagram 

The heatmap visualization provides a correlation matrix that shows the relationships between the different variables in the dataset. Here's an interpretation of the data and a recommendation for columns to drop if the target variable is the ghgScore:

### Interpretation:

### Correlation Analysis:

Strong negative correlations with ghgScore: Variables like co2 and comb appear to have a strong inverse relationship with ghgScore, meaning as these values increase, the ghg score decreases.
Strong positive correlations with ghgScore: Variables like feScore (fuel efficiency score) have a strong positive correlation with ghgScore, indicating that vehicles with better fuel efficiency have higher ghg scores.

### Weak or no correlations:

Some variables, such as id, year, pv2, and pv4, seem to have little to no correlation with ghgScore, suggesting they do not contribute significantly to explaining its variation.

### Other significant relationships:

Variables like displ and cylinders likely have an indirect impact on ghgScore through their relationships with fuel consumption metrics (comb, city, etc.).


### Columns to Drop if targeting ghgScore:

id: It is just an identifier and does not contribute to the target variable.
year: The correlation with ghgScore is negligible, and it might not provide predictive value in this context.
pv2 and pv4: These variables likely represent categorical or dummy variables for specific conditions (e.g., passenger vehicle types). Their low correlation with ghgScore makes them candidates for removal unless they have significant categorical effects.
Variables highly correlated with other predictors: Variables like highway, UHighway, city, and UCity might be redundant, as they are highly correlated with comb (combined fuel consumption). Including them could introduce multicollinearity.

### Justification:
Dropping these columns reduces noise and redundancy in the dataset, focusing the model on features that have stronger, more direct relationships with the target variable (ghgScore). Additionally, removing highly correlated predictors (e.g., comb vs. city vs. highway) minimizes multicollinearity, improving model stability and interpretability.
